```python
df_country=pd.read_excel('Documents\Python Scripts\Country-Code.xlsx')
df_country.head()
```

```python
df_final=pd.merge(df,df_country,on='Country Code', how='left')
df_final.head()
```

```python
df_final.dtypes
```

```python
Country_Names= df_final.Country.value_counts().index
Country_Names
```

Observation: India is the largest user of Zomato. 


```python
Country_Values= df_final.Country.value_counts().values
```

```python
#Represetation of Top 3 Countries that uses Zomato via Pie Chart
plt.pie(Country_Values[0:3], labels=Country_Names[0:3], autopct='%1.2f%%')
```

Observation1: From the above diagram, it is evident that the maximum users of Zomato are in India followed by US and Uk. Therefore we can conclude that highest business for Zomato is generated by India


```python
df_final.columns
```

```python
result=df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
result
```

```python
#fig = plt.figure(figsize = (10, 5))
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating count", data=result)
plt.xlabel("Rating")
plt.ylabel("Rating count")
plt.title("Zomato Rating")
plt.show()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating count",hue="Rating color", data=result, palette=['blue','red','orange','yellow','green','green'])
plt.xlabel("Rating")
plt.ylabel("Rating count")
plt.title("Zomato Rating")
plt.show()
```

Observation
1. Not rated count is very high
